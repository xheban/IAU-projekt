Všimli sme si, že atribút personal_info v sebe ukrýva osobné údaje - ďalšie atribúty. Chceme ho rozložiť, aby sme s ním vedeli pracovať.

```python
def process_pi(text):
    if pd.isnull(text):
        return pd.Series([np.nan]*5)
    splitted_pi = re.split('[| \r\r\n]',text)
    splitted_pi = list(filter(None,splitted_pi))
    splitted_pi.remove("--")
    return pd.Series([splitted_pi[0],splitted_pi[1],splitted_pi[2],splitted_pi[3],splitted_pi[4]])
```

Funkcia process_pi vie rozdeliť reťazec na jednotlivé atribúty podľa oddeľujúcich znakov.

```python
other_train[['job','state','family_status','home','color']] = other_train['personal_info'].apply(process_pi)
```

## Hypotézy

### Hypotéza 1
#### Nulová hypotéza:
Ľudia s cukrovkou majú rovnakú hladinu cukru v krvi ako ľudia bez cukrovky.
#### Alternatívna hypotéza:
Ľudia s cukrovkou majú zvýšenú/zníženú hladinu cukru v krvi.



#### Začneme
Z datasetu si vytiahneme hodnoty priemernej hladiny cukru v krvi zvlášť pre ľudí s a bez cukrovky.

```python
hladina_cukru_s_cukrovkou = other_train.loc[other_train['class'] == 1,'mean_glucose'].dropna()
hladina_cukru_bez_cukrovky = other_train.loc[other_train['class'] == 0,'mean_glucose'].dropna()
```

Opis hodnôt pre oba prípady.

```python
print("Hodnoty pre ľudí bez cukrovky")
hladina_cukru_bez_cukrovky.describe()
```

```python
print("Hodnoty pre ľudí s cukrovkou")
hladina_cukru_s_cukrovkou.describe()
```

Už tu si môžeme všimnúť, že priemerná hodnota cukru v krvi pre ľudí trpiacich cukrovkou je o dosť nižšia ako u ľudí netrpiacich cukrovkou. Skúsme to ešte raz po tom, ako odstránime outlierov, ktorých si najskôr vypočítame.

```python
lowQ,highQ = hladina_cukru_s_cukrovkou.quantile([0.25, 0.75])
interquartile = highQ-lowQ
max = highQ + 1.5*(interquartile)
min = lowQ - 1.5*(interquartile)
print("Boxplot:\n *minimum value: ",min,"\n *maximum value",max)
tmp = hladina_cukru_s_cukrovkou[~hladina_cukru_s_cukrovkou.between(min,max)]
pocet = tmp.count();
print("Počet outlierov: ", pocet)
sb.boxplot(hladina_cukru_s_cukrovkou).set_title("Cukor u cukrovkárov")
print("Nemusíme vylúčiť žiadnych outlierov, pretože v danej sade žiadní nie sú")
```

```python
lowQ,highQ = hladina_cukru_bez_cukrovky.quantile([0.25, 0.75])
interquartile = highQ-lowQ
max = highQ + 1.5*(interquartile)
min = lowQ - 1.5*(interquartile)
print("Boxplot:\n *minimum value: ",min,"\n *maximum value",max)
tmp = hladina_cukru_bez_cukrovky[~hladina_cukru_bez_cukrovky.between(min,max)]
pocet = tmp.count();
print("Počet outlierov: ", pocet)
sb.boxplot(hladina_cukru_bez_cukrovky).set_title("Cukor u ne-cukrovkárov s outliermi")
print("Vylúčime si 23 outleirov")

def delete_out(record):
    return record[ (record > max) | (record < min)]
```

## 2. fáza - Predspracovanie

```python
personal_train = pd.read_csv("personal_train.csv",sep=',', index_col = 0)
```

```python
def repair_age(age):
    if pd.isnull(age):
        return pd.Series(np.nan)
    if age >= 130:
        return age / 100
    if age < 0:
        return ((age * -1) / 100)
    else:
        return age
```

```python
personal_train['age'] = personal_train['age'].apply(repair_age)
```
